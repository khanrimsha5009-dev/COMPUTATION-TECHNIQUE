```python
-score test (simple)
# Given: population mean = 85, claimed/sample mean = 88, sigma = 1
# NOTE: sample size (n) is not provided in the question, so we assume n = 1.
# If your teacher gave an n, replace n below.

import math

mu = 85 # population average (% good mangoes)
x_bar = 88 # person's garden (% good mangoes)
sigma = 1 # given
n = 1 # assumed (change if n is given)
alpha = 0.05 # common significance level

# Z statistic
z = (x_bar - mu) / (sigma / math.sqrt(n))

# Critical value for two-tailed test at alpha=0.05 (â‰ˆ 1.96)
z_crit = 1.96

decision = "REJECT H0 (claim not accepted)" if abs(z) > z_crit else "ACCEPT H0 (claim accepted)"

print("Z =", z)
print("Critical Z (two-tailed, alpha=0.05) =", z_crit)
print("Decision:", decision)
```
