```python
#Q1)# Q1
TP = int(input("Enter first number: "))
TN = int(input("Enter second number: "))
FP = int(input("Enter third number: "))
FN = int(input("Enter fourth number: "))

total = TP + TN + FP + FN

accuracy = (TP + TN) / total
precision = TP / (TP + FP)
recall = TP / (TP + FN)
f1_score = 2 * (precision * recall) / (precision + recall)

print("accuracy:", accuracy)
print("precision:", precision)
print("recall:", recall)
print("f1_score:", f1_score)

```
